## Композиція  в об'єктно-орієнтованому програмуванні  

Композиція це створення об'єктів існуючих класів (реальних об'єктів) як елементів інших класів (класів програми)

Композиція - це важливіша концепція ООП, яка є визначальною при проектуванні моделі даних і створенню відношень між класами програми.


```python
class Figure:
    color = "червоний"
    form = "круг"
    def changecolor(self,newcolor):
        self.color = newcolor
    def changeform(self,newform):
        self.form = newform

obj1 = Figure()
obj2 = Figure()

print (obj1.color, obj1.form) # "червоний круг"
print (obj2.color, obj2.form) # "червоний круг"

a = obj1 == obj2
print(a)

obj1.changecolor("зелений") # зміна кольору obj1 
obj2.changecolor("синій") # зміна кольору obj2
obj2.changeform("еліпс") # зміна форми obj2

print (obj1.color, obj1.form) # "зелений круг"
print (obj2.color, obj2.form) # "синій еліпс"
```

```python
class YesConstructor:
    def __init__(self,one ,two):
        self.fname = one
        self.sname = two

obj1 = YesConstructor("Призвище","Ім'я")


print (obj1.fname, obj1.sname)

class NoConstructor:
    def names(self,one,two):
        self.fname = one
        self.sname = two

obj2 = NoConstructor()

# очікуваний результат
# print (obj2.fname, obj2.sname)


obj2.names("Призвище_1","Ім'я_1")
# звернення до методу класу, якщо немає конструктору 

print (obj2.fname, obj2.sname)

# очікуваний результат
obj3 = YesConstructor()

print(obj3)
```

```python
# значення атрибутів класу, які обов'язково вказуються при створенні екземплярів
# класу, за замовчуванням, вказуються першими в констурукторі класу
class fruits:
    def __init__(self,w,n=0):
        self.what = w
        self.numbers = n

f1 = fruits("яблуко",150)
f2 = fruits("лимон")

print (f1.what,f1.numbers)
print (f2.what,f2.numbers)
```

```python
# інтервальні атрибути класу, операції зі значенніми атрибутів класу
class Warehouse:
    def __init__(self,w,c,n=0):
        self.what = w
        self.color = c
        self.numbers = n
        self.mwhere(n)

# умовне значення атрибуту mwhere(n)  
    def mwhere(self,n):
        if n <= 0:
            self.where = ": відсутня"
        elif 0 < n < 100:
            self.where = ": промисловий склад"
        else:
            self.where = ": оптовий склад"

    def plus(self,p):
        self.numbers = self.numbers + p
        self.mwhere(self.numbers)
        
    def minus(self,m):
        self.numbers = self.numbers - m
        self.mwhere(self.numbers)

m1 = Warehouse("Цегла", "червона",1500)
m2 = Warehouse("Сайдинг", "коричневий", 300)
m3 = Warehouse("Фарба","біла")

print (m1.what,m1.color,m1.where)
print (m2.what,m2.color,m2.where)
print (m3.what,m3.color,m3.where)

m1.plus(500)
print (m1.numbers, m1.where)

m2.minus(200)
print (m2.numbers, m2.where)
```

```python
class Table:
    def __init__(self,l,w,h):
        self.long = l
        self.width = w
        self.height = h
    def outparam(self):
        print (self.long,self.width,self.height)

class Kitchen(Table):   
    def howplaces(self,n):
        if n < 2:
            print ("Це не кухонний стіл")
        else:
            self.places = n
    def outplases(self):
        print (self.places)

t_room1 = Kitchen(2,1,0.5)
t_room1.outparam()
t_room1.howplaces(1)
t_room1.outplases()

t_room2 = Table(1,3,0.7)
t_room2.outparam()
# прогнозований результат
t_room2.howplaces(8) # 
```

```python
# поліморфізм (багато форм) - метод total 
class T1:
    n=10
    def total(self,N):
        self.total = int(self.n) + int(N)

class T2:
    def total(self,s):
        self.total = len(str(s))

t1 = T1()
t2 = T2()
t1.total(238)
t2.total(238)
print (t1.total) #  238
print (t2.total) #  3
```

```python
# перевизначення методів базового класу успадкованими класами 
class Base:
    def __init__(self,n):
        self.numb = n
    def out(self):
        print (self.numb)

class One(Base):
    def multi(self,m):
        self.numb *= m

class Two(Base):
    def inlist(self):
        self.inlist = list(str(self.numb))
    def out(self):
        i = 0
        while i < len(self.inlist):
            print (self.inlist[i])
            i += 1

obj1 = One(50)
obj2 = Two('АВСД')

obj1.multi(2)
obj1.out() #  100

# прогнозований результат 
obj1.inlist()

obj2.inlist()
obj2.out() #  А, Б, С, Д
```

```python
# розширення методів базового класу (за умови)
class Base:
    def __init__(self,N):
        self.numb = N
    def out(self):
        self.numb /= 2   
        
class Subclass(Base):
    def out(self):
        Base.out(self)

i = 0
while i < 10:
    if 4 < i < 7:
        obj = Subclass(i)
        print ("\n----SubClass")
    else:
        obj = Base(i)
        print ("BaseClass----\n")
    i += 1
obj.out()
```
